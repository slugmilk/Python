# **1. 튜플**
여러 가지 데이터 타입의 값을 저장할 수 있는, 변경 불가능(immutable)한 순서가 있는 컬렉션입니다

```python
tu1 = (1, 3, 5, 7)
print(tu1)
print(type(tu1))
```

```python
tu1 = (1, 3, 5, 7)
tu1[4] = 9
```

```python
tu2 = (10)
print(tu2)
print(type(tu2)) # 괄호 안에 하나의 요소만 넣으면 튜플로 인식 X. 연산자 ()로 인식
```

```python
# 튜플 속 요소가 한 개인 경우 끝에 , 넣어줘야 함
tu2 = (10,)
print(tu2)
print(type(tu2))
```

```python
tu3 = tuple([1, 3, 5, 7]) # 함수로도 튜플 생성 가능
print(tu3)
print(type(tu3))
```

```python
tu4 = 1, 3, 5, 7 # 괄호 없이도 만들 수 있음
print(tu4)
print(type(tu4))
```

```python
tu5 = ('apple', 'banna', ('🍎', '🍌')) # 튜플 안에 튜플 넣을 수 있음
print(tu5)
print(type(tu5))
print(tu5[0])
print(type(tu5[0]))
print(tu5[2][0])
print(type(tu5[2][0]))

print(type(tu5[2]))
```

```python
tu5 = ('apple', 'banna', ('🍎', '🍌'))
#tu5[0] =  'orange'
#TypeError: 'tuple' object does not support item assignment
print(tu5)
```

```python
tu6 = ('apple', 'banna', ['🍎', '🍌'])
print(tu6)
print(tu6[2][0])
tu6[2][0] = '🍋' # 튜플 속 리스트의 항목을 바꾸는 건 됨
print(tu6)

# tu6[2] = '🍒'
# TypeError: 'tuple' object does not support item assignment
print(type(tu6[2]))
```

```python
tu7 = (10, 20, 30)
tu8 = (40, 50, 60)

print(tu7 + tu8) # 덧셈 가능
print(tu8 + tu7)

tu7 = tu7 + (40, 50) # 다른 요소를 더하는 건 주소가 바뀌는 것이기 때문에 가능
print(tu7)

print(tu7 * 3)
```

```python
tu7 = (10, 20, 30)
print(tu7[0] + tu7[2])
```

```python
apple, banna, orange = ('김사과', '반하나', '오렌지') # 괄호나 대괄호로 묶어서 한 번에 변수 선언 가능
print(apple)
print(banna)
print(orange)
```

```python
apple, banna, orange = '김사과', '반하나', '오렌지' # 사실 안묶어도 됨 에헷!
print(apple)
print(banna)
print(orange)
```

```python
print(2 in (2, 4, 6, 8, 10)) # 뒤의 튜플 안에 요소가 있는지?
print(3 in (2, 4, 6, 8, 10))
```

```python
tu8 = (100, 30, 70, 20, 90)
# tu8.sort()
# AttributeError: 'tuple' object has no attribute 'sort'
# 튜플은 요소를 변경할 수 없기 때문에 inplace X
print(sorted(tu8)) # sorted는 list 반환
type(sorted(tu8))
```

# **2. input() 함수**
input() 함수는 사용자로부터 키보드 입력을 받는 데 사용되는 내장 함수입니다. 프로그램 실행 중에 사용자가 값을 입력하면, 입력한 값이 문자열로 반환되어 변수에 저장하거나 처리할 수 있습니다.

```python
name = input()
```

```python
print(name)
```

```python
name = input('이름을 입력하세요: ')
```

```python
print(f'{name}님 안녕하세요!')
```

```python
# 유민콩 4 15
# '유민콩 4 15' -> ['유민콩', '4', '15']
name, age, weight = input('이름, 나이, 몸무게를 컴마를 기준으로 입력하세요').split(',')
```

```python
print(name)
print(age)
print(weight)
```

```python
num1 = input('첫번째 숫자를 입력하세요: ')
num2 = input('두번째 숫자를 입력하세요: ')
print('결과: ', num1 + num2)
```

```python
# int(): 문자열을 정수형으로 변환하는 함수
num1 = input('첫번째 숫자를 입력하세요: ')
num2 = input('두번째 숫자를 입력하세요: ')
print('결과: ', int(num1) + int(num2))
```

### 문제
* 국어점수, 영어점수, 수학점수를 입력받아 총점, 평균을 구하는 프로그램을 작성해보자
* 단, 입력방법은 input 함수를 이용하여 100/70/85 형태로 입력

```python
kor, eng, math = input('국어점수/영어점수/수학점수를 입력하세요: ').split('/')
total = int(kor) + int(eng) + int(math)
average = total / 3
print(total, average)
```
