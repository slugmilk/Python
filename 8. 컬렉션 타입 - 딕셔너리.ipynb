# **1. 딕셔너리**
* 파이썬의 딕셔너리는 키-값 쌍을 저장하는 변경 가능한(mutable) 컬렉션입니다.
(키는 변경 불가하고 값을 변경할 수 있다)
* 순서X
* 키는 중복될 수 없지만 값은 중복될 수 있다 (키는 set으로 이루어져 있기 때문)


```python
dict1 = {}
print(dict1)
print(type(dict1))
```

```python
dict2 = {1:'유민콩', 2:'정현콩', 3:'수현콩', 4:'유빈콩'}
print(dict2)
print(type(dict2))
```

```python
print(dict2[2])
print(dict2[4]) # 인덱스 아님
```

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}
print(dict3)
print(dict3['userid'])
print(dict3['name'])
```

```python
dict4 = {1:'유민콩'}
print(dict4)

dict4[100] = '유빈콩' # 메서드 없이 값 추가 가능
print(dict4)

dict4[50] = '정현콩'
print(dict4)

dict4[100] = '수현콩' # 같은 키에 값 넣으면 그 값으로 변경됨
print(dict4)
```

```python
del dict4[100] # 키 제거
print(dict4)
```

# **2. 키, 값의 제약**

딕셔너리의 키는 변경 불가능한(immutable) 타입이어야 합니다. 예를 들어, 문자열, 정수, 튜플은 딕셔너리의 키로 사용할 수 있지만, 리스트는 딕셔너리의 키로 사용할 수 없습니다. 하지만 딕셔너리의 값은 어떤 타입이든 상관없습니다.

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}
print(dict3)

dict3['no'] = 10 # 값 변경
print(dict3)

dict3['gender'] = 'f' # 키-값 추가
print(dict3)

dict3['score'] = [90, 100, 60] # 값을 리스트 형태로도 저장 가능
print(dict3)

# dict3[[1, 2, 3]] = ['하나', '둘', '셋']
# TypeError: unhashable type: 'list'
# 키에는 리스트 사용 불가

dict3[(1, 2, 3)] = ['하나', '둘', '셋'] # 키에 튜플은 사용 가능
print(dict3)

dict3['fruit'] = {'사과':'🍎', '딸기':'🍓', '수박':'🍉'} # 딕셔너리에 딕셔너리 저장 가능
print(dict3)
```

# **3. 딕셔너리의 메서드**

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}
print(len(dict3)) # 키-값이 몇 개 들어있는지
```

### 3-1. keys(): 딕셔너리의 모든 키를 반환

```python
print(dict3.keys())
```

### 3-2. values(): 딕셔너리의 모든 값을 반환

```python
print(dict3.values())
```

### 3-3. items(): 딕셔너리의 모든 키-값을 쌍(튜플)으로 반환

```python
print(dict3.items())
```

### 3-4. get(): 특정 키에 대한 값을 반환. 만약 키가 딕셔너리에 없으면 None을 반환

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}
print(dict3['userid'])
```

```python
# print(dict3['gender'])
# KeyError: 'gender'
# 없는 키값 부르면 오류
```

```python
print(dict3.get('userid'))
print(dict3.get('gender')) # None
print(dict3.get('age', '나이를 알 수 없음')) # None을 대신해서 출력할 문장
print(dict3.get('userid', '아이디를 알 수 없음'))
```

### 3-5. pop(): 특정 키에 대한 값을 제거하고 그 값을 반환. 키가 없다면 에러

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}

temp = dict3.pop('hp')
print(dict3)
print(temp)
```

### in: 딕셔너리의 특정 키가 있는지 확인 (메서드X)

```python
dict3 = {'no':1, 'userid':'yoomin', 'name':'유민콩', 'hp':'010-1234-5678'}
print('hp' in dict3)
print('gender' in dict3)
```
