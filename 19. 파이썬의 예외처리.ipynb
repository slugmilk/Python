# **4. ì‚¬ìš©ì ì •ì˜ ì˜ˆì™¸ í´ë˜ìŠ¤ë¥¼ ì§ì ‘ ë§Œë“¤ê³  í™œìš©í•˜ê¸°**

````python
# 20ì‚´ ë¯¸ë§Œ: ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ ì–´ë¦¼
# 20ì‚´ ~ 50ì‚´: ë‚˜ì´ ë²”ìœ„ ì•ˆì— í¬í•¨
# 50ì‚´ ì´ˆê³¼: ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ í¼

class AgeLimitError(Exception):
  def __init__(self, age, message='ì›í•˜ëŠ” ë‚˜ì´ ë²”ìœ„ê°€ ì•„ë‹˜'): # ë©”ì„¸ì§€ì˜ ê¸°ë³¸ê°’ ì„¤ì •í•¨
    self.age = age
    self.message = message
    super().__init__(self.message) # Exception í´ë˜ìŠ¤ì— ë©”ì„¸ì§€ ë³´ëƒ„ (Exception(message))
````

````python
def check_age(age):
  if age < 20:
    raise AgeLimitError(age, 'ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ ì–´ë¦¼')
  elif age > 50:
    raise AgeLimitError(age, 'ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ í¼')
  else:
    return 'ë‚˜ì´ ë²”ìœ„ ì•ˆì— í¬í•¨'
````

````python
ages = [17, 60, 46, 20, 52, 26]

for age in ages:
  try:
    print(check_age(age))
  except AgeLimitError as e:
    print(f'{e.age} ë‚˜ì´ ì—ëŸ¬: {e}')
````
