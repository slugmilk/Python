# **1. 스페셜 메서드**
파이썬의 스페셜 메서드 (또는 매직 메서드라고도 불림)는 더블 언더스코어(__)로 시작하고 끝나는 메서드 이름을 갖습니다. 이 메서드들은 특정 구문이나 내장 함수를 사용할 때 파이썬 인터프리터에 의해 *자동으로* 호출됩니다.

```python
class Book:
  def __init__(self, title):
    self.title = title
```

```python
book = Book('어떻게 햄스터가 개발자')
print(book)
print(str(book)) # print 안에 str()이 생략됐다고 볼 수 있다. str()은 문자열로 변환해주는 기능을 갖지만, 객체를 넣으면 객체의 정보를 보여준다
```

```python
class Book:
  def __init__(self, title):
    self.title = title
  def __str__(self): # 파이썬 내의 str()함수를 오버라이딩 (이 클래스 한정)
    return self.title
```

```python
book = Book('어떻게 햄스터가 개발자')
print(book)
print(str(book))
```

```python
li = [1, 2, 3]
print(li) # li는 객체인데 정보가 아니라 안의 데이터가 출력됨. 즉 list의 클래스의 __str__에 리스트 안의 요소를 출력할 수 있도록 오버라이딩 돼있음
print(str(li))
```

```python
class Calc:
  def __init__(self, num):
    self.num = num

  def __add__(self, other): # + 기호를 오버라이딩
    return self.num + other.num
```

```python
num1 = Calc(5)
num2 = Calc(10)
print(num1 + num2) # 객체의 요소 간 덧셈이 가능하게 함
```

```python
class Queue:
  def __init__(self):
    self.items = [1, 2, 3, 4, 5]

  def __len__(self): # len() 함수 오버라이딩
    return len(self.items)
```

```python
li = [1, 2, 3, 4, 5]
print(len(li))
```

```python
queue = Queue()
print(queue) # str()을 오버라이딩 시키지 않아서 정보가 나옴
print(len(queue)) # len() 함수를 객체에서도 쓸 수 있게 함
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __getitem__(self, index): # 인덱싱할 때 호출(인덱싱 오버라이딩)
    if index == 0:
      return self.x
    elif index == 1:
      return self.y
    else:
      return -1
```

```python
pt = Point(5, 3)
print(pt[0])
print(pt[1])
print(pt[2])
```
