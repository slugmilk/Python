### 7-9. ìŠ¬ë¼ì´ì‹±ì„ ì‚¬ìš©í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì˜ ìˆœì„œë¥¼ ë’¤ì§‘ëŠ” ë°©ë²•
ìŠ¬ë¼ì´ì‹±(slicing)ì€ íŒŒì´ì¬ì—ì„œ ì‹œí€€ìŠ¤ íƒ€ì…ì˜ ê°ì²´(ì˜ˆ: ë¦¬ìŠ¤íŠ¸, ë¬¸ìì—´, íŠœí”Œ ë“±)ì˜ ì¼ë¶€ë¥¼ ì¶”ì¶œí•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. ìŠ¬ë¼ì´ì‹±ì˜ ê¸°ë³¸ êµ¬ì¡°ëŠ” [start:stop:step]ì…ë‹ˆë‹¤.

1. start: ìŠ¬ë¼ì´ìŠ¤ì˜ ì‹œì‘ ì¸ë±ìŠ¤ì…ë‹ˆë‹¤. ì´ ì¸ë±ìŠ¤ì˜ ìš”ì†ŒëŠ” ìŠ¬ë¼ì´ìŠ¤ì— í¬í•¨ë©ë‹ˆë‹¤. startë¥¼ ìƒëµí•˜ë©´, stepì´ ì–‘ìˆ˜ì´ë©´ 0, stepì´ ìŒìˆ˜ì´ë©´ -1ë¡œ ê°„ì£¼ë©ë‹ˆë‹¤.

2. stop: ìŠ¬ë¼ì´ìŠ¤ì˜ ë ì¸ë±ìŠ¤ì…ë‹ˆë‹¤. ì´ ì¸ë±ìŠ¤ì˜ ìš”ì†ŒëŠ” ìŠ¬ë¼ì´ìŠ¤ì— í¬í•¨ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤. stopì„ ìƒëµí•˜ë©´, stepì´ ì–‘ìˆ˜ì´ë©´ ì‹œí€€ìŠ¤ì˜ ë, stepì´ ìŒìˆ˜ì´ë©´ ì‹œí€€ìŠ¤ì˜ ì‹œì‘ìœ¼ë¡œ ê°„ì£¼ë©ë‹ˆë‹¤.

3. step: ìŠ¬ë¼ì´ìŠ¤ì˜ ìŠ¤í…(ì¦ê°) ê°’ì…ë‹ˆë‹¤. stepì„ ìƒëµí•˜ë©´ 1ë¡œ ê°„ì£¼ë©ë‹ˆë‹¤.

````python
li8 = [100, 50, 80, 60, 20]
print(li8)

print(li8[0:5]) # ë¦¬ìŠ¤íŠ¸0ë²ˆë¶€í„° 5ë²ˆ ì§ì „ê¹Œì§€
print(li8[0:5:1])
print(li8[:5:1])
print(li8[::1])
print(li8[::])

print(li8[::-1]) # ìŠ¬ë¼ì´ì‹±ìœ¼ë¡œ ë¦¬ìŠ¤íŠ¸ ë’¤ì§‘ê¸° ê°€ëŠ¥
````
