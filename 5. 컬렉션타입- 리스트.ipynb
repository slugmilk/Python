# **4. 리스트의 슬라이싱**

````python
li1 = [1, 3, 5, 7, 9]
print(li1[0:4:2])
print(li1)
print(li1[0:3])
print(li1[:2])
print(li1[1:])
print(li1[:])
````

````python
li7 = [10, 20, 30, ['김사과', '오렌지', '반하나'], 40, ['💟', '💢']]
print(li7[2:6])
print(li7[5][:1]) #와웅 헷갈리는걸
````

````python
li1 = [1, 3, 5, 7, 9]
print(li1[0]) #인덱스로 뽑으면 무조건 차원이 하나 날라감 -> 대괄호 하나 날라감
print(li1[0:1]) #슬라이싱 하면 차원이 유지됨 대괄호 고대로 갖고 나옴
````

````python
li6 = [1, 2, 3, ['유민콩', '정현콩', '수현콩', ['🍉', '🍓', '🍎', '🍒']]]
print(li6[3][3][0])
print(li6[3][3:4][0:1]) #인덱싱 하나여서 대괄호 하나 날라감 [3:4]은 슬라이싱으로 뽑아서 대괄호 유지
````

````python
#인덱싱을 이용하여 요소를 추가한 경우 리스트 안에 리스트를 만들고 포함(차원이 증가)
li1 = [1, 3, 5, 7, 9]
li1[1]=['😊', '😂', '👍']
print(li1) #인덱싱으로 넣으면 차원이 늘어나서 대괄호 추가
````

````python
#슬라이싱을 이요하여 요소를 추가한 경우 리스트에 데이터만 포함(차원이 유지)
li1 = [1, 3, 5, 7, 9]
li1[1:2] = ['😊', '😂', '👍']
print(li1) #슬라이싱으로 넣으면 차원 그대로
````

# **7. 리스트의 함수와 메서드**



### 7-1. len() : 객체의 길이(항목의 개수)를 반환하는 내장 함수

````python
li1 = [1, 3, 5, 7, 9]

print(len(li1)) #요소의 개수 출력
````

### 7-2. appennd() : 리스트에 사용되며, 리스트의 끝에 새로운 항목을 하나만 추가하는 메서드

````python
li1 = [1, 3, 5, 7, 9]
print(li1)

li1.append(100)
print(li1)

#li1.append(200, 300, 400) #TypeError: list.append() takes exactly one argument (3 given)
li1.append([200, 300, 400]) #묶어서 하나로 들어감
print(li1)
````

### 7-3. extend() : 리스트에 iterable(반복 가능한 객체)의 모든 항목을 추가하는 메서드

````python
li1 = [1, 3, 5, 7, 9]
print(li1)

li1.extend([50])
print(li1)

li1.extend([100, 200])
print(li1)
````

### 7-4. pop() : 리스트에서 항목을 삭제하고, 삭제된 항목을 반환하는 메서드

````python
li1 = [1, 3, 5, 7, 9]
print(li1)

print(li1.pop()) #빠진거 알려쥼
print(li1)

temp = li1.pop()  #빼고 temp에 넣어쥼
print(temp)
print(li1)
````

### 7-5. insert() : 리스트의 특정 인덱스에 항목을 추가하는 메서드

````python
li1 = [1, 3, 5, 7, 9]
li1.insert(1, 100) #1번 인덱스 자리에 100 추가 후 나머지 애들 뒤로 보냄
print(li1) # 출력:
li1.insert(2, [1000, 2000])
print(li1) # 출력:
````

### 7-7. reverse(): 리스트의 항목들의 순서를 뒤집는 메소드

````python
li8 = [100, 50, 80, 60, 20]
print(li8)
li8.reverse() # 반대로 저장. in-place 연산(자기 자신에게 다시 재저장)
print(li8)
````

### 7-8. sort(): 리스트의 항목들을 정렬하는 메소드. inplace 연산을 수행

````python
li8 = [100, 50, 80, 60, 20]
print(li8)
li8.sort() # 오름차순
print(li8)
````

````python
li8 = [100, 50, 80, 60, 20]
print(li8)
li8.sort(reverse = True) # 내림차순
print(li8)
````

````python
li9 = ['Apple', 'apple', 'orange', 'Banna', 'banna', 'melon']
li9.sort()
print(li9) # 첫 번째 문자의 아스키코드 기준으로 오름차순
````

````python
li10 = ['유민콩', '유빈콩', '정현콩', '수현콩']
li10.sort()
print(li10) # 유니코드
````

💡비교
* sorted(): iterable(반복 가능한, 순서가 있는 객체)의 모든 항목들을 정렬한 후, 그 결과를 새로운 리스트에 담아 변환하는 함수

````python
li8 = [100, 50, 80, 60, 20]
print(li8)

print(sorted(li8))
print(li8) # 원래대로 돌아옴. inplace X

result = sorted(li8)
print(result)
````

````python
li8 = [100, 50, 80, 60, 20]
print(li8)

result = sorted(li8, reverse = True) # 내림차순
print(result)
````

### 7-9. 슬라이싱을 사용하여 리스트의 순서를 뒤집는 방법
슬라이싱(slicing)은 파이썬에서 시퀀스 타입의 객체(예: 리스트, 문자열, 튜플 등)의 일부를 추출하는 방법입니다. 슬라이싱의 기본 구조는 [start:stop:step]입니다.

1. start: 슬라이스의 시작 인덱스입니다. 이 인덱스의 요소는 슬라이스에 포함됩니다. start를 생략하면, step이 양수이면 0, step이 음수이면 -1로 간주됩니다.

2. stop: 슬라이스의 끝 인덱스입니다. 이 인덱스의 요소는 슬라이스에 포함되지 않습니다. stop을 생략하면, step이 양수이면 시퀀스의 끝, step이 음수이면 시퀀스의 시작으로 간주됩니다.

3. step: 슬라이스의 스텝(증감) 값입니다. step을 생략하면 1로 간주됩니다.

````python
li8 = [100, 50, 80, 60, 20]
print(li8)

print(li8[0:5]) # 리스트0번부터 5번 직전까지
print(li8[0:5:1])
print(li8[:5:1])
print(li8[::1])
print(li8[::])

print(li8[::-1]) # 슬라이싱으로 리스트 뒤집기 가능
````

### 7-10. count(): 리스트에서 특정 값의 개수를 반환하는 메소드
