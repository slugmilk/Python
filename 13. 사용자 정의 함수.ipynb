# **1. 사용자 정의 함수**
사용자 정의 함수란 사용자가 특정 작업을 수행하기 위해 직접 작성한 함수를 의미합니다. 파이썬에는 많은 내장 함수들이 있지만, 때로는 우리의 요구사항에 맞게 동작하는 함수를 직접 만들어야 할 때가 있습니다. 이때 사용자 정의 함수를 작성하게 됩니다. 기본적인 구조는 아래와 같습니다.
```
def 함수명(매개변수1, 매개변수2, ...):
    # 함수 내용
    return 결과값
```

### 1-1. 매개변수와 반환 값이 없는 함수

````python
def func1():
    print('🐹: 찍찍')
````

````python
func1()
func1()
func1()
````

````python
for i in range(5):
    func1()
````

### 1-2. 매개변수가 있고 반환 값이 없는 함수

````python
def func2(num):
    print(f'입력받은 숫자: {num}')
````

````python
func2(100)
func2(5)
````

````python
def func3(start, end):
    sum = 0
    for i in range(start, end+1):
        sum += i
    print(f'{start}부터 {end}까지의 합: {sum}')
````

````python
func3(1, 10)
func3(1, 100)
````

### 1-3. 반환 값이 있는 함수

````python
def func4():
    print('안녕하세요 Python!')
````

````python
func4()
temp = func4()
print(temp) # 함수에서 실행된 값이 프린트 되고 사라져서 temp가 가리킬 값이 없어짐
````

````python
temp = func4 # 이건 실행이 아니라 함수의 메모리 주소 저장
print(temp)
````

````python
temp()
````

````python
def func5():
    return '🐹'
````

````python
print(func5()) # 출력된 후 사라짐
````

````python
temp = func5()
print(f'temp에 저장된 값: {temp}')
````

````python
def func6(num1, num2):
    sum = num1 + num2
    return sum
````

````python
print(func6(10, 5)) # 출력하고 가리킬 게 없어 사라짐
temp = func6(10, 5) # temp에 값 저장됨
print(temp)
````

### 1-4. 기본값이 설정된 매개변수


````python
def func6(num1=0, num2=0):
  sum = num1 + num2
  return sum
````

````python
print(func6())
print(func6(10))
print(func6(10, 3))
# print(func6(, 3))
# SyntaxError: invalid syntax
# print(func6(None, 3))
# TypeError: unsupported operand type(s) for +: 'NoneType' and 'int'
print(func6(num2=3))
````

### 1-5. 가변 매개변수
함수를 호출할 때 *를 사용하면 시퀀스(리스트, 튜플 등)의 요소르르 개별적인 위치 인자로 풀어서 전달할 수 있다.

````python
def func7(*args):
  return args
````

````python
print(func7()) # 빈 튜플 생성
print(func7(10, 30))
print(func7(10, 30, 70, 90, 100)) # 튜플의 요소 수 제약없이 값 변경 가능
````

### 1-6. 키워드 매개변수

````python
def func8(id, name, age):
  print(f'아이디: {id}')
  print(f'이름: {name}')
  print(f'나이: {age}')
````

````python
func8('yoomin', '유민콩', 5)
````

````python
func8(age=5, id='yoomin', name='유민콩') # 매개변수만 맞으면 순서 상관 X
````

````python
dic1 = {'age':5, 'id':'yoomin', 'name':'유민콩'}
func8(**dic1) # 딕셔너리로 넣을 땐 앞에 **
````

````python
func8(*dic1) # *이 하나 빠지면 키가 값으로 전달
````

### 1-7. 여러개의 변환값

````python
def func9(num1, num2):
  return num1 + num2
````

````python
result1 = func9(10, 3)
print(result1)
````

````python
def func10(num1, num2):
  return num1+num2, num1-num2, num1*num2, num1/num2 # 여러개 반환 가능
````

````python
result1, result2, result3, result4 = func10(10, 3) # 여러 변수에 차례로 대입 가능
print(f'덧셈: {result1}')
print(f'뺄셈: {result2}')
print(f'곱셈: {result3}')
print(f'나눗셈: {result4}')
````

````python
_, _, result3, _ = func10(10, 3) # 사용하지 않을 변수를 만들고 싶지 않을때 (메모리 절약을 위해) _를 통해 특정 변수를 만들지 않을 수 있음
print(f'곱셈: {result3}')
````

````python
li1 = [10, 20, 30, 40, 50]

for _, v in enumerate(li1):
  print(f'값: {v}')
````

# **2. None의 특징**
None은 파이썬에서 특볋한 값으로, 아무런 값이 없음을 표현하는 데 사용된다. 다른 언어에서의 null또는 nill과 유사한 개념이다. None은 파이썬의 내장 상수이며, 그 자체로 데이터 타입이 NoneType이다. 모든 None는 동일하므로, 두 개의 None값을 비교할 때 항상 True를 반환한다.

````python
num = None
print(num)
````

````python
def func11(num):
  if num >= 10:
    return num
````

````python
print(func11(10))
print(func11(4))
````

````python
def func12(num=None):
  if num is None:
    print('전달된 값이 없습니다.')
  else:
    print(num)
````

````python
func12(10)
func12()
````

````python
num = None

if num is None:
  print('num의 값은 없습니다.')

# == 을 통해 None을 비교하는 것은 권장하지 않음
# None은 가리키는 메모리 자체가 없다는 의미이기 때문
if num == None:
  print('num의 값은 없습니다.')
````

````python
li1 = [1, 2, 3, 4, 5]
li2 = [1, 2, 3, 4, 5]

# li1과 li2의 메모리 주소가 동일한가? (객체 비교)
print(li1 is li2)
# li1과 li2가 동일한가? (값 비)
print(li1 == li2)
````
