# **1. print 함수**
print() 함수는 파이썬의 기본 출력 함수로, 터미널이나 콘솔 창에 텍스트나 데이터를 출력하기 위해 사용됩니다.


```python
print('안녕하쇼')
print(100)
print('10')
print(10, 20, 30, 40)
print(10, '이십', 30, '40')
```

```python
print(10, 20, 30, 40)
print('개굴', '개굴', '개굴', '개굴', sep='🐸', end='🐸\n') # sep= print문 사이에 들어갈 구분자. 기본은 띄어쓰기
print('개굴', '개굴', '개굴', '개굴', sep='🐸', end='🐸')
```

```python
print('Hello', 'Python')
print('Hello', end=' ') # end=의 기본값은 엔터임
print('Python')
print('Hello Python')
```

# **2. 파이썬의 escape 문자**
파이썬에서 이스케이프 문자는 문자열 안에서 특별한 의미를 가지는 문자들을 나타내기 위해 사용됩니다. 이스케이프 문자는 백슬래시(\)로 시작하며, 문자열 내에서 특정한 동작이나 특수문자를 표현하기 위해 사용됩니다.

```python
print('Hello', 'Python', end='\n')
print('Hello\nPython')
```

```python
print('안녕하세요\n파이썬')
print('이름:\t유민콩') # \t는 탭키 누른 것 같은 효과
print('파이썬은 너무 재밌?네 \\') # 화면에 백슬러쉬 나타내고 싶을 떄 \\를 사용
print("'안녕하세요'") # 작은 따움표까지 출력하고 싶을 땐 큰 따움표로 감싸주기. 반대의 경우도 반대로 하면 됨
print('"안녕하세요"')
print("\"안녕하세요\"") # 굳이굳이 같은 따움표 안에 따움표를 출력하고 싶으면 역슬래시 붙이면 됨.. 근데 굳이?
```

# **3. print 함수 내의 연산**
print() 함수는 문자열뿐만 아니라 숫자와 연산 결과도 출력할 수 있습니다.

```python
print('덧셈: ', 10 + 8)
print('뺄셈: ', 10 - 8)
print('곱셈: ', 10 * 8)
print('나눗셈: ', 10 / 2) # *파이썬은 나눗셈을 하면 결과가 정수라도 실수로 나옴*
print('정수 나눗셈: ', 10 // 5) # 몫만 구함 --> 정수로 출력
print('나머지: ', 10 % 3) # 나머지만 구함
print('거듭제곱: ', 10**3)
```

# **4. 출력서식 지정**
파이썬의 print() 함수는 출력 서식을 지정하여 값을 형식화하고 출력하는 기능을 제공합니다. 서식 지정자와 함꼐 사용되는 % 연산자와 format() 매서드를 사용하여 원하는 형식으로 출력할 수 있습니다.

```python
print('정수: %d' % 100) # %d -> % 뒤에 오는 숫자(정수)를 %d 자리에 집어넣음
print('정수: %d' % 100.5) # 실수로 넣어도 정수로 보여줌
```

```python
print('실수: %f' % 100) # %f -> 소수 여섯번째 자리까지 출력
print('실수: %f' % 100.5)
```

```python
print('문자열: %s' % 100) # %s -> 문자열 출력. 숫자를 넣어도 문자열로 출력됨
print('문자열: %s' % 'Python')
```

```python
print(f'정수: {100}') # f-string. 중괄호 안에 변수명을 넣어서 사용할 수 있다.
```

```python
print('5 + 3 = 8')
print('%d + %d = %d' % (5, 3, 5+3)) # ()안의 값을 차례대로 넣어줌
```

```python
print('아름: %s, 나이: %d' % ('유민콩', 5)) # 자료형이 달라도 사용 O
print('몸무게: %f' % 13.12345)
print('몸무게: %.2f' % 13.12345) # %.2f -> 소숫점 둘쨰자리까지 반올림
print('내친구: {} {}'.format('정현콩', '수현콩')) #.format()은 문자열에만 사용 가능한 메소드
```

# **5. 주석문**

```python
print('파이썬 출력하기') # 문자열 출력은 작은 따옴표 또는 큰 따옴표를 사용한다
```

```python
'''
이곳은 여러 줄 주석입니다.
여러 줄 주석은 큰 따옴표 또는 작은 따옴표 세 개 사이에 작상합니다.
여러 줄에 걸쳐 설명을 적을 수 있습니다
'''
print('응애')
```
